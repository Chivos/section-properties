```python
from sectionproperties.pre import sections
from sectionproperties.analysis.cross_section import Section

from matplotlib.pyplot import plot
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
```

```python
rhs = sections.rectangular_hollow_section(b=200, d=100, t=10, r_out=20, n_r=12)
```

```python
rhs.create_mesh([3])
sec = Section(rhs)
sec.calculate_geometric_properties()
sec.calculate_warping_properties()
sec.calculate_plastic_properties()
```

```python
stress_results = sec.calculate_stress(Mzz=100e3)
```

```python
fig, ax = stress_results.plot_stress_mzz_zxy(normalize=False, size=2000, dpi=200)
```
