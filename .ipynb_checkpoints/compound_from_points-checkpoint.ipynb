```python
import sectionproperties.pre.sections as sections
from sectionproperties.analysis.cross_section import Section
from sectionproperties.pre.pre import Material
```

```python
mat1 = Material(name="Material 1", elastic_modulus=100, poissons_ratio=0.3, yield_strength=10, density=1000, color="yellow")
mat2 = Material(name="Material 2", elastic_modulus=100, poissons_ratio=0.3, yield_strength=10, density=1000, color="orange")
mat3 = Material(name="Material 3", elastic_modulus=100, poissons_ratio=0.3, yield_strength=10, density=1000, color="red")

sq1 = sections.rectangular_section(100, 100, material=mat1).align_center()
sq2 = sections.rectangular_section(75, 75, material=mat2).align_center()
sq3 = sections.rectangular_section(50, 50, material=mat3).align_center()
hole = sections.rectangular_section(25, 25).align_center()

compound = (sq1 - sq2) + (sq2 - sq3) + (sq3 - hole)
compound
```

```python
v1_points = compound.points
v1_facets = compound.facets[0:8] + compound.facets[12:16] + compound.facets[20:]
v1_ctrl = compound.control_points
v1_holes = [compound.holes[0]]
```

```python
comp2 = sections.CompoundGeometry.from_points(points=v1_points, facets=v1_facets, control_points=v1_ctrl, holes=v1_holes)
```

```python
comp2.plot_geometry()
```

```python
comp2.create_mesh([10, 5, 2.5])
```

```python
sec2 = Section(comp2)
```

```python
from sectionproperties.pre import sections
from sectionproperties.pre.pre import Material
from sectionproperties.analysis.cross_section import Section
```

```python
mat1 = Material("Steel", 200, 0.3, 300, 1000, "red")
mat2 = Material("OtherSteel", 200, 0.3, 400, 1000, "grey")

d = 300
b = 150
tf = 10
tw = 8
r = 12
b_p = 250
t_p = 16
ub = sections.i_section(d=d, b=b, t_f=tf, t_w=tw, r=r, n_r=16, material=mat1)
plate = sections.rectangular_section(b=b_p, d=t_p, material=mat2).align_center(ub).align_to(ub, on="top")
geom = ub + plate
geom = geom.offset_perimeter(amount=5, where='exterior')
# geom.plot_geometry()
geom.create_mesh([100])
section = Section(geom)
section.plot_mesh()
```

```python
geom.create_mesh([10])
```

```python
uc = sections.i_section(d=400,b=400,t_f=25,t_w=25,r=30,n_r=8)
plate1 = sections.rectangular_section(b=500,d=10).align_center(uc).align_to(uc,"top")
plate2 = sections.rectangular_section(b=500,d=10).align_center(uc).align_to(uc,"bottom")
geom = uc + plate1 + plate2

for i in range(15):
    new_geom = geom.offset_perimeter(-i)
    display(new_geom)
```

```python
uc = sections.i_section(d=400,b=400,t_f=25,t_w=25,r=30,n_r=8)
plate1 = sections.rectangular_section(b=500,d=10).align_center(uc).align_to(uc,"top")
plate2 = sections.rectangular_section(b=500,d=10).align_center(uc).align_to(uc,"bottom")
geom = uc + plate1 + plate2

new_geom = geom.offset_perimeter(-9)
new_geom.create_mesh([100])
section = Section(new_geom)

print("PENULATIMATE: \n")
new_geom = geom.offset_perimeter(-10)
print(new_geom.control_points)
new_geom.create_mesh([100])

section = Section(new_geom)

print("FINAL: \n")
new_geom = geom.offset_perimeter(-11)
new_geom.create_mesh([100])
section = Section(new_geom)
```
