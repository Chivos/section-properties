```python
import sectionproperties.pre.sections as sections
from sectionproperties.pre.pre import Material
from sectionproperties.analysis.cross_section import Section
```

```python
mat1 = Material("Material_1", 200e3, 0.3, 400, "green")
mat2 = Material("Material_2", 150e3, 0.2, 200, "red")

top_flange = sections.rectangular_section(250, 20).align_center()
web = top_flange.rotate_section(90).align_center(top_flange).align_to(top_flange, on='bottom')
bottom_flange = top_flange.align_center(web).align_to(web, on='bottom')

simple_i_sec = top_flange | web | bottom_flange
simple_i_sec.material = mat1

rot_i_sec = simple_i_sec.rotate_section(45)
rot_i_sec.material = mat2
```

```python
cut_2_from_1 = (simple_i_sec - rot_i_sec) # locates intersection nodes
sec_1_nodes_added = cut_2_from_1 | simple_i_sec
composite = sec_1_nodes_added + (rot_i_sec - simple_i_sec)
composite.create_mesh([100])
an_sec = Section(composite)
an_sec.plot_mesh()
```

```python
an_sec.calculate_geometric_properties()
an_sec.calculate_warping_properties()
stress = an_sec.calculate_stress(N=100e3, M22 = 100e3, )
```
